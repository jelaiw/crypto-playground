```python
from cryptography.hazmat.primitives import poly1305
```

```python
import os
```

```python
k = os.urandom(32)
```

```python
p = poly1305.Poly1305(k)
```

```python
p.update(b'message')
```

```python
tag = p.finalize()
```

```python
from binascii import hexlify as hexa
```

```python
print(hexa(tag))
```
